### Given a board with clean spots denoted as '-' and dirty spots denoted as 'd' move a bot 'b' to clean the board

```python
from math import sqrt
# location of the bot b
posr = 0
posc = 0
# board
board = ['-d--d','-d--d','--dd-','--d--','----d']
# assigning bot
r = list(board[posr])
r[posc] = 'b'
r=''.join(r)
board[posr]=r
board

# for ib  in board:
#     print(ib)
```

```python
def dirty_dis(dirty_loc,curr_loc):
    diff= list(map(lambda x:[x[0]-curr_loc[0],x[1]-curr_loc[1]],dirty_loc))
    y = list(map(lambda x: sqrt(x[0]**2+x[1]**2),diff))
    return y
def what2do(target_loc,curr_pos):
    x = target_loc[0]-curr_pos[0]
    y = target_loc[1]-curr_pos[1]

    if x>0:
        new_dir =  'DOWN'
    elif x<0:
        new_dir =  'UP'

    if x==0 and y>0:
        new_dir =  'RIGHT'
    elif x==0 and y<0:
        new_dir =  'LEFT'
    elif x==0 and y==0:
        new_dir =  'CLEAN'


    return new_dir     

def stepupdate(curr_pos,new_dir):
    if new_dir =='DOWN':
        curr_pos[0] +=1
        CC=True
    if new_dir =='UP':
        curr_pos[0] -=1
        CC=True

    if new_dir =='LEFT':
        curr_pos[1] -=1
        CC=True
    if new_dir =='RIGHT':
        curr_pos[1] +=1
        CC=True
    if new_dir=='CLEAN':
        CC = False

    return curr_pos, CC   

def board_plotter(board,curr_loc,next_loc):
    row_board = list(board[curr_loc[0]])
    row_board[curr_loc[1]] = '-'
    board[curr_loc[0]] = ''.join(row_board)

    row_board2 = list(board[next_loc[0]])
    row_board2[next_loc[1]] = 'b'
    board[next_loc[0]] = ''.join(row_board2)
    for i in board:
        print(i)

curr_loc = [posr,posc]
dirty_loc = []
for i in range(len(board)):
    for j in range(len(board[i])):
        if board[i][j] == 'd':
            dirty_loc.append([i,j])
# for i in board:
#     print(i)
for i in range(len(dirty_loc)):
# for i in range(2):

#     print('current location:',curr_loc)
    d = dirty_dis(dirty_loc,curr_loc)
    order_dis = sorted(range(len(d)),key=d.__getitem__)
    target = dirty_loc[order_dis[0]]
#     print('target:',dirty_loc[order_dis[0]])
    CC = True
    while CC:
#         
        c = curr_loc.copy()
        new_dir = what2do(target,c)
        print(new_dir)
        next_loc,CC = stepupdate(c,new_dir)

        board_plotter(board,curr_loc,next_loc)
        curr_loc = next_loc.copy()
        print('__________________________________________________')
        if not CC:
            dirty_loc.pop(order_dis[0])
```
