### Given a board with clean spots denoted as '-' and dirty spots denoted as 'd' move a bot 'b' to clean the board

```python
import pygame
white = (255,255,255)
black=(0,0,0)
red = (255,0,0)


w = 100
wbot = 20
wl=2
def board_draw(board,screen,w,wbot,wl):
    pygame.event.get()
    xci,yci = int(w/2),int(w/2) 
    xsqi,ysqi = w/2 - wbot/2,w/2 - wbot/2 
    x,y = 0,0
    for row in board:
        for col in row:
#             print('('+col+')')
            pygame.draw.rect(screen,white,[x,y,w,w])
            pygame.display.update()
            if col =='b':
#                 print('b:',col)
#                 print(xci)
#                 print(yci)
                pygame.draw.circle(screen,red,[xci,yci],int(wbot),10)
                pygame.display.update()
            elif col =='d':
#                 print('d:',col)
                pygame.draw.rect(screen,black,[xsqi,ysqi,wbot,wbot])
                pygame.display.update()


            x = x+w+wl
            xci = xci + w
            xsqi = xsqi + w
                ##############################
        yci = yci+w
        ysqi = ysqi+w
        y = y+w+wl

        xci = int(w/2) 
        xsqi = w/2 - wbot/2
        x = 0
        

        
```

```python
from math import sqrt
import time
# location of the bot b
posr = 3
posc = 1
# board
# board = ['-d--d-d-','-d--d--d','--dd---d','--d--dd-','dddddddd']
board = ['-d','-d','--','--','dd']

# assigning bot
r = list(board[posr])
r[posc] = 'b'
r=''.join(r)
board[posr]=r
board
screen = pygame.display.set_mode((w*len(board[0]),w*len(board)))
screen.fill(black)
board_draw(board,screen,w,wbot,wl)

# for ib  in board:
#     print(ib)
```

```python
def dirty_dis(dirty_loc,curr_loc):
    diff= list(map(lambda x:[x[0]-curr_loc[0],x[1]-curr_loc[1]],dirty_loc))
    y = list(map(lambda x: sqrt(x[0]**2+x[1]**2),diff))
    return y
def what2do(target_loc,curr_pos):
    x = target_loc[0]-curr_pos[0]
    y = target_loc[1]-curr_pos[1]

    if x>0:
        new_dir =  'DOWN'
    elif x<0:
        new_dir =  'UP'

    if x==0 and y>0:
        new_dir =  'RIGHT'
    elif x==0 and y<0:
        new_dir =  'LEFT'
    elif x==0 and y==0:
        new_dir =  'CLEAN'


    return new_dir     

def stepupdate(curr_pos,new_dir):
    if new_dir =='DOWN':
        curr_pos[0] +=1
        CC=True
    if new_dir =='UP':
        curr_pos[0] -=1
        CC=True

    if new_dir =='LEFT':
        curr_pos[1] -=1
        CC=True
    if new_dir =='RIGHT':
        curr_pos[1] +=1
        CC=True
    if new_dir=='CLEAN':
        CC = False

    return curr_pos, CC   

def board_plotter(board,curr_loc,next_loc):
    row_board = list(board[curr_loc[0]])
    row_board[curr_loc[1]] = '-'
    board[curr_loc[0]] = ''.join(row_board)

    row_board2 = list(board[next_loc[0]])
    row_board2[next_loc[1]] = 'b'
    board[next_loc[0]] = ''.join(row_board2)
#     for i in board:
#         print(i)
    return board

curr_loc = [posr,posc]
dirty_loc = []
for i in range(len(board)):
    for j in range(len(board[i])):
        if board[i][j] == 'd':
            dirty_loc.append([i,j])
# for i in board:
#     print(i)
for i in range(len(dirty_loc)):
# for i in range(2):

#     print('current location:',curr_loc)
    d = dirty_dis(dirty_loc,curr_loc)
    order_dis = sorted(range(len(d)),key=d.__getitem__)
    target = dirty_loc[order_dis[0]]
#     print('target:',dirty_loc[order_dis[0]])
    CC = True
    while CC:
#         
        c = curr_loc.copy()
        new_dir = what2do(target,c)
#         print(new_dir)
        next_loc,CC = stepupdate(c,new_dir)
        board = board_plotter(board,curr_loc,next_loc)

#         print(board)
        board_draw(board,screen,w,wbot,wl)
        time.sleep(1)
        curr_loc = next_loc.copy()
#         print('__________________________________________________')
        if not CC:
            dirty_loc.pop(order_dis[0])
```
